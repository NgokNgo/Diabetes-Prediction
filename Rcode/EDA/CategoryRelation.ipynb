```r
library(ggplot2)
library(dplyr)
library(readr)
library(tidyr)
library(reshape2)
```

```r
setwd("") # set your working directory

df <- read_csv('diabetes_012_health_indicators_BRFSS2015.csv')
head(df)
```

```r
df_category <- df %>% select(-BMI, -MentHlth, -PhysHlth, -Diabetes_012)
```

```r
par(bg = 'white')
for (i in 1:ncol(df_category)) {
  for (j in (i+1):ncol(df_category)) {
    if (i != j && i != ncol(df_category) && j != ncol(df_category)) {
        mosaicplot(table(as.vector(as.matrix(df_category[,i])), as.vector(as.matrix(df_category[,j]))), 
          main = paste("Mosaic plot for", colnames(df_category)[i], "and", colnames(df_category)[j]), 
          color = TRUE, shade = TRUE)
    }
  }
}
```
